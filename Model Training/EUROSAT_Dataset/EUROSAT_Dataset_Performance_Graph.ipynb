```python
import json
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#reading RESNET performance graph 
with open('/content/drive/MyDrive/COMP6721_GROUP_N/EuroSat Dataset Weights/ACC and Loss values/EuroSat_dataset_results_ResNet.txt', 'r') as f:
    eurosat_dataset_results_resnet = json.loads(f.read())

#reading VGG performance graph 
with open('/content/drive/MyDrive/COMP6721_GROUP_N/EuroSat Dataset Weights/ACC and Loss values/EuroSat_dataset_results_vgg.txt', 'r') as f:
    eurosat_dataset_results_vgg = json.loads(f.read())

#reading ENET performance graph 
with open('/content/drive/MyDrive/COMP6721_GROUP_N/EuroSat Dataset Weights/ACC and Loss values/EuroSAT_dataset_results_enet.txt', 'r') as f:
    eurosat_dataset_results_enet = json.loads(f.read())
```

**Plotting Accuracy of EUROSAT Dataset with all three models - VGG16, ResNet50 and ENet_b0**

```python
num_epochs = 40
plt.figure(figsize=(10,8))

#vgg16 model
plt.plot(np.arange(1, num_epochs+1),
         eurosat_dataset_results_vgg['train_acc_list']+ [None]*(40-len(eurosat_dataset_results_vgg['train_acc_list'])), label='VGG16 Training')
plt.plot(np.arange(1, num_epochs+1),
          eurosat_dataset_results_vgg['valid_acc_list']+ [None]*(40-len(eurosat_dataset_results_vgg['valid_acc_list'])), label='VGG16 Validation')

# resnet50 Model
plt.plot(np.arange(1, num_epochs+1),
          eurosat_dataset_results_resnet['train_acc_list']+ [None]*(40-len(eurosat_dataset_results_resnet['train_acc_list'])), label='ResNet50 Training')
plt.plot(np.arange(1, num_epochs+1),
          eurosat_dataset_results_resnet['valid_acc_list']+ [None]*(40-len(eurosat_dataset_results_resnet['valid_acc_list'])), label='ResNet50 Validation')

#enet_b0
plt.plot(np.arange(1, num_epochs+1),
          eurosat_dataset_results_enet['train_acc_list'], label='ENet_b0 Training')
plt.plot(np.arange(1, num_epochs+1),
          eurosat_dataset_results_enet['valid_acc_list'], label='ENet_b0 Validation')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.title("Train VS Validation ACCURACY of EUROSAT Dataset with VGG16, ResNet50 and Enet_b0")
plt.tight_layout()
```

```python
num_epochs = 40
plt.figure(figsize=(10,8))

#vgg16 model
plt.plot(np.arange(1, num_epochs+1),
          eurosat_dataset_results_vgg['train_loss_list']+ [None]*(num_epochs-len(eurosat_dataset_results_vgg['train_loss_list'])), label='VGG16 Training')
plt.plot(np.arange(1, num_epochs+1),
          eurosat_dataset_results_vgg['valid_loss_list']+ [None]*(num_epochs-len(eurosat_dataset_results_vgg['valid_loss_list'])), label='VGG16 Validation')

# resnet50 Model
plt.plot(np.arange(1, num_epochs+1),
          eurosat_dataset_results_resnet['train_loss_list']+ [None]*(num_epochs-len(eurosat_dataset_results_resnet['train_loss_list'])), label='ResNet50 Training')
plt.plot(np.arange(1, num_epochs+1),
         eurosat_dataset_results_resnet['valid_loss_list']+ [None]*(num_epochs-len(eurosat_dataset_results_resnet['valid_loss_list'])), label='ResNet50 Validation')

#enet_b0
plt.plot(np.arange(1, num_epochs+1),
          eurosat_dataset_results_enet['train_loss_list'], label='ENet_b0 Training')
plt.plot(np.arange(1, num_epochs+1),
          eurosat_dataset_results_enet['valid_loss_list'], label='ENet_b0 Validation')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.title("Train VS Validation LOSS of EUROSAT Dataset with VGG16, ResNet50 and Enet_b0")
plt.tight_layout()
```
