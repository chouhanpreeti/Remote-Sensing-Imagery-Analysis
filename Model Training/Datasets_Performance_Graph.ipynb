```python
import json
import matplotlib.pyplot as plt 
import numpy as np
```

**Reading Satellite Dataset accuracy and loss**

```python
#reading RESNET performance graph 
with open('/content/drive/MyDrive/COMP6721_GROUP_N/Satellite Dataset Weights/ACC and Loss values/Satellite_dataset_results_resnet.txt', 'r') as f:
    Satellite_dataset_results_resnet = json.loads(f.read())

#reading VGG performance graph 
with open('/content/drive/MyDrive/COMP6721_GROUP_N/Satellite Dataset Weights/ACC and Loss values/Satellite_dataset_results_vgg.txt', 'r') as f:
    Satellite_dataset_results_vgg = json.loads(f.read())

#reading ENET performance graph 
with open('/content/drive/MyDrive/COMP6721_GROUP_N/Satellite Dataset Weights/ACC and Loss values/Satellite_dataset_results_enet.txt', 'r') as f:
    Satellite_dataset_results_enet = json.loads(f.read())
```

**Reading EUROSAT Dataset accuracy and loss**

```python
#reading RESNET performance graph 
with open('/content/drive/MyDrive/COMP6721_GROUP_N/EuroSat Dataset Weights/ACC and Loss values/EuroSat_dataset_results_ResNet.txt', 'r') as f:
    eurosat_dataset_results_resnet = json.loads(f.read())

#reading VGG performance graph 
with open('/content/drive/MyDrive/COMP6721_GROUP_N/EuroSat Dataset Weights/ACC and Loss values/EuroSat_dataset_results_vgg.txt', 'r') as f:
    eurosat_dataset_results_vgg = json.loads(f.read())

#reading ENET performance graph 
with open('/content/drive/MyDrive/COMP6721_GROUP_N/EuroSat Dataset Weights/ACC and Loss values/EuroSAT_dataset_results_enet.txt', 'r') as f:
    eurosat_dataset_results_enet = json.loads(f.read())
```

**Reading Land-Use Dataset accuracy and loss**

```python
#reading RESNET performance graph 
with open('/content/drive/MyDrive/COMP6721_GROUP_N/LandUse Dataset Weights/ACC and Loss values/landuse_dataset_resnet_results.txt', 'r') as f:
    LandUse_dataset_results_resnet = json.loads(f.read())

#reading VGG performance graph 
with open('/content/drive/MyDrive/COMP6721_GROUP_N/LandUse Dataset Weights/ACC and Loss values/landuse_dataset_vgg_results.txt', 'r') as f:
    LandUse_dataset_results_vgg = json.loads(f.read())

#reading ENET performance graph 
with open('/content/drive/MyDrive/COMP6721_GROUP_N/LandUse Dataset Weights/ACC and Loss values/landuse_dataset_enet_results.txt', 'r') as f:
    LandUse_dataset_results_enet = json.loads(f.read())
```

**Plotting Validation Accuracy of Satellite, EUROSAT and Land-Use Dataset with all three models - VGG16, ResNet50 and ENet_b0**

```python
num_epochs = 40
plt.figure(figsize=(10,8))

#vgg16 model
plt.plot(np.arange(1, num_epochs+1),
          eurosat_dataset_results_vgg['valid_acc_list']+ [None]*(40-len(eurosat_dataset_results_vgg['valid_acc_list'])), label='EuroSat - VGG16')
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg['valid_acc_list']+ [None]*(40-len(LandUse_dataset_results_vgg['valid_acc_list'])), label='Land-Use - VGG16')
plt.plot(np.arange(1, num_epochs+1),
          Satellite_dataset_results_vgg['valid_acc_list'], label='Satellite - VGG16')

# resnet50 Model
plt.plot(np.arange(1, num_epochs+1),
          eurosat_dataset_results_resnet['valid_acc_list']+ [None]*(40-len(eurosat_dataset_results_resnet['valid_acc_list'])), label='EuroSat - ResNet50')
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_resnet['valid_acc_list']+ [None]*(40-len(LandUse_dataset_results_resnet['valid_acc_list'])), label='Land-Use - ResNet50')
plt.plot(np.arange(1, num_epochs+1),
          Satellite_dataset_results_resnet['valid_acc_list']+ [None]*(40-len(Satellite_dataset_results_resnet['valid_acc_list'])), label='Satellite - ResNet50')

#enet_b0
plt.plot(np.arange(1, num_epochs+1), eurosat_dataset_results_enet['valid_acc_list'], label='EuroSat - ENet_b0')
plt.plot(np.arange(1, num_epochs+1), LandUse_dataset_results_enet['valid_acc_list'], label='Land-Use - ENet_b0')
plt.plot(np.arange(1, num_epochs+1),
          Satellite_dataset_results_enet['valid_acc_list'], label='Satellite - ENet_b0')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.title("Validation ACCURACY of Satellite, EUROSAT and Land-Use Dataset with VGG16, ResNet50 and Enet_b0")
plt.tight_layout()
```

**Plotting Validation Loss of Satellite, EUROSAT and Land-Use Dataset with all three models - VGG16, ResNet50 and ENet_b0**

```python
num_epochs = 40
plt.figure(figsize=(10,8))

#vgg16 model
plt.plot(np.arange(1, num_epochs+1),
          eurosat_dataset_results_vgg['valid_loss_list']+ [None]*(num_epochs-len(eurosat_dataset_results_vgg['valid_loss_list'])), label='EuroSat - VGG16')
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg['valid_loss_list']+ [None]*(num_epochs-len(LandUse_dataset_results_vgg['valid_loss_list'])), label='Land-Use - VGG16')
plt.plot(np.arange(1, num_epochs+1),
          Satellite_dataset_results_vgg['valid_loss_list'], label='Satellite - VGG16')

# resnet50 Model
plt.plot(np.arange(1, num_epochs+1),
         eurosat_dataset_results_resnet['valid_loss_list']+ [None]*(num_epochs-len(eurosat_dataset_results_resnet['valid_loss_list'])), label='EuroSat - ResNet50')
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_resnet['valid_loss_list']+ [None]*(num_epochs-len(LandUse_dataset_results_resnet['valid_loss_list'])), label='Land-Use - ResNet50')
plt.plot(np.arange(1, num_epochs+1),
          Satellite_dataset_results_resnet['valid_loss_list']+ [None]*(num_epochs-len(Satellite_dataset_results_resnet['valid_loss_list'])), label='Satellite - ResNe50')


#enet_b0
plt.plot(np.arange(1, num_epochs+1), eurosat_dataset_results_enet['valid_loss_list'], label='EuroSat - ENet_b0')
plt.plot(np.arange(1, num_epochs+1), LandUse_dataset_results_enet['valid_loss_list'], label='Land-Use - ENet_b0')
plt.plot(np.arange(1, num_epochs+1), Satellite_dataset_results_enet['valid_loss_list'], label='Satellite - ENet_b0')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.title("Validation LOSS of Satellite, EUROSAT and Land-Use Dataset with VGG16, ResNet50 and Enet_b0")
plt.tight_layout()
```
