```python
import json
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#reading VGG1 performance graph 
with open('/content/drive/MyDrive/AI Project/LandUseResults/landuse_dataset_vgg1results.txt', 'r') as f:
    LandUse_dataset_results_vgg1 = json.loads(f.read())

#reading VGG2 performance graph 
with open('/content/drive/MyDrive/AI Project/LandUseResults/landuse_dataset_vgg2results.txt', 'r') as f:
    LandUse_dataset_results_vgg2 = json.loads(f.read())

#reading VGG3 performance graph 
with open('/content/drive/MyDrive/AI Project/LandUseResults/landuse_dataset_vgg3results.txt', 'r') as f:
    LandUse_dataset_results_vgg3 = json.loads(f.read())
```

**Plotting Accuracy of LandUse Dataset with all three models - VGG16, ResNet50 and ENet_b0**

```python
num_epochs = 40
plt.figure(figsize=(10,8))

#vgg16 Model 1
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg1['train_acc_list']+ [None]*(40-len(LandUse_dataset_results_vgg1['train_acc_list'])), label='VGG16 Training1')
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg1['valid_acc_list']+ [None]*(40-len(LandUse_dataset_results_vgg1['valid_acc_list'])), label='VGG16 Validation1')

# vgg16 Model 2
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg2['train_acc_list']+ [None]*(40-len(LandUse_dataset_results_vgg2['train_acc_list'])), label='VGG16 Training2')
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg2['valid_acc_list']+ [None]*(40-len(LandUse_dataset_results_vgg2['valid_acc_list'])), label='VGG16 Validation2')

#vgg16 Model 3
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg3['train_acc_list']+ [None]*(40-len(LandUse_dataset_results_vgg3['train_acc_list'])), label='VGG16 Training3')
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg3['valid_acc_list']+ [None]*(40-len(LandUse_dataset_results_vgg3['valid_acc_list'])), label='VGG16 Validation3')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.title("Train VS Validation ACCURACY of LandUse Dataset with various VGG Trainings")
plt.tight_layout()
```

```python
num_epochs = 40
plt.figure(figsize=(10,8))

#vgg16 model
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg1['train_loss_list']+ [None]*(num_epochs-len(LandUse_dataset_results_vgg1['train_loss_list'])), label='VGG16 Training1')
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg1['valid_loss_list']+ [None]*(num_epochs-len(LandUse_dataset_results_vgg1['valid_loss_list'])), label='VGG16 Validation1')

# resnet50 Model
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg2['train_loss_list']+ [None]*(num_epochs-len(LandUse_dataset_results_vgg2['train_loss_list'])), label='VGG16 Training2')
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg2['valid_loss_list']+ [None]*(num_epochs-len(LandUse_dataset_results_vgg2['valid_loss_list'])), label='VGG16 Validation2')

#enet_b0
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg3['train_loss_list']+ [None]*(num_epochs-len(LandUse_dataset_results_vgg3['train_loss_list'])), label='VGG16 Training3')
plt.plot(np.arange(1, num_epochs+1),
          LandUse_dataset_results_vgg3['valid_loss_list']+ [None]*(num_epochs-len(LandUse_dataset_results_vgg3['valid_loss_list'])), label='VGG16 Validation3')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.title("Train VS Validation LOSS of LandUse Dataset with VGG16, ResNet50 and Enet_b0")
plt.tight_layout()
```
